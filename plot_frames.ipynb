```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import flowtracks
from flowtracks import io
```

```python
for tr in io.iter_trajectories_ptvis('./test_data/ptv_is.%d'):
    print(tr.time()[-1])
```

```python
max_frame = 101010
trajectories = []
for tr in io.iter_trajectories_ptvis('./test_data/ptv_is.%d'):
    if tr.time()[-1] <= max_frame:
        trajectories.append(tr)
```

```python
len(trajectories)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for tr in trajectories:
    plt.plot(tr.pos()[:,0],tr.pos()[:,1])
    plt.plot(tr.pos()[0,0],tr.pos()[0,1],'.')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for traj in trajectories:
	ax.plot(traj.pos()[:,0], traj.pos()[:,1], traj.pos()[:,2], '.')

plt.show()
```
